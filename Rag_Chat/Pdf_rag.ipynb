```python
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader("Resistance.pdf")
file=loader.load()
```

```python
file
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=40)
documents=text_splitter.split_documents(file)
documents[:2]
```

```python
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.vectorstores import FAISS
db=FAISS.from_documents(documents,OllamaEmbeddings())
```

```python
query="Eccentrics are better than concentrics"
result=db.similarity_search(query)
result[0].page_content
```

```python
from langchain_community.llms import Ollama
llm=Ollama(model='llama2')
```

```python
from langchain_core.prompts import ChatPromptTemplate
prompt=ChatPromptTemplate.from_template(
    """Answer Based on the provided context
    <context>
    {context}
    </context>
    Question: {input}"""
)
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain
document_chain=create_stuff_documents_chain(llm,prompt)
```

```python
retreiver=db.as_retriever()
```

```python
from langchain.chains import create_retrieval_chain
retrieval_chain=create_retrieval_chain(retreiver,document_chain)
```

```python
response=retrieval_chain.invoke({"Eccentrics are better than concentrics"})
```
